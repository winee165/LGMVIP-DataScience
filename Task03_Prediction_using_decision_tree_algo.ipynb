# PREDICTION USING DECISION TREE  ALGORITHM

Importing Libraries

```python
import numpy as np
import pandas as pd
```

Load and Read Dataset

```python
df=pd.read_csv("/content/Iris.csv")
df.head()
```

```python
x=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=df[['Species']]
```

Importing DecisionTreeClassifier Library and fitting data in Model

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x,y)
```

Importing Libraries for Data Visualisation

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

Visualising Model as Graph

```python
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names =['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'], filled = True, rounded = True, special_characters = True,proportion=False, node_ids = True)
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
